```python
df = pd.read_csv('SOCR-HeightWeight.csv')
df
```

```python
mean = df['Height(Inches)'].mean()
mean
```

```python
std = df['Height(Inches)'].std()
std
```

```python
up = mean + 3* std
up
```

```python
low = mean - 3* std
low
```

```python
df[df['Height(Inches)']<low]
```

```python
df[df['Height(Inches)']>up]
```

```python
df[(df['Height(Inches)']>up) | (df['Height(Inches)']<low)]
```

```python
df[(df['Height(Inches)']>up) & (df['Height(Inches)']<low)]
```

```python
df['zScore'] = (df['Height(Inches)']-mean)/std
df
```

```python

outliers = df[(df['zScore'] > 3) | (df['zScore'] < -3)]


outliers
```

```python
min = df['Height(Inches)'].min()

min
```

```python
max = df['Height(Inches)'].max()
max
```

```python
df['meanHeight'] = (df['Height(Inches)'] - min) / (max - min)
df
```

```python
wmin = df['Weight(Pounds)'].min()

wmin
```

```python
wmax = df['Weight(Pounds)'].max()
wmax
```

```python
df['meanWeight'] = (df['Weight(Pounds)'] - wmin) / (wmax - wmin)
df
```

```python

def fibonacci_series(n):

  a, b = 0, 1
  for i in range(n):
    print(a, end=" ")
    a, b = b, a + b


fibonacci_series(10)
```
