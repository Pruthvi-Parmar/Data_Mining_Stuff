```python
company = ['google','microsoft',"amazon",'isro']
revenue = [100,200,300,400]
profit = [10,20,30,40]
plt.xlabel('company')
plt.ylabel('revenue')
plt.title('company analysis')
plt.bar(company,revenue,color = 'lightblue')
plt.bar(company,profit,color = 'lightcoral')
import numpy as np
xpos = np.arange(len(company))
plt.xticks(xpos,company)
```

```python
plt.xlabel('company')
plt.ylabel('revenue')
plt.title('company analysis')
plt.bar(xpos-0.2,revenue,width=0.4)
plt.bar(xpos+0.2,profit,width=0.4)
plt.xticks(xpos,company)
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ['home','rent','bills','car','other']
plt.pie(exp_vals,labels = exp_labels,shadow=True,colors = ['lightblue','lightcoral','lightgreen','pink','yellow'],autopct='%.2f%%',explode=(0,0,0,0,0.5))
plt.title('Expense Report')
plt.show()
```

```python
import seaborn as sn
df = sn.load_dataset('tips')
df
```

```python
plt.scatter(df['total_bill'],df['tip'],color = 'lightcoral')
```

```python
plt.scatter(df['total_bill'][:20], df['tip'][:20], color='lightcoral')
plt.show()
```

```python
plt.hist(df['total_bill'],color = 'lightcoral',width=3)
plt.xlabel('total_bill')
plt.ylabel('frequency')
plt.show()
```

```python
sn.boxplot(x = df['sex'],y = df['total_bill'],width=.2,color='lightblue')
```

```python
days = df['day'].unique()
day_wise_total_bill = []
for day in days:
    day_wise_total_bill.append(df[df['day']==day]['total_bill'].sum())
plt.bar(days,day_wise_total_bill,color = 'lightcoral',width=0.5)
plt.xlabel('Days')
plt.ylabel('Total Bill')
plt.title('Day wise total bill')
plt.show()
```

```python
smoker_count = df['smoker'].value_counts()
plt.pie(smoker_count, labels=smoker_count.index, colors=['lightcoral', 'lightblue'], autopct='%.2f%%')
plt.show()
```
